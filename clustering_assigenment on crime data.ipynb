```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
crime=pd.read_csv("Crime.csv")
crime
```

```python
crime.describe
```

```python
crime.info()
```

#normalising the given data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
norm=MinMaxScaler()
norm
```

```python
data=pd.DataFrame(norm.fit_transform(crime.iloc[:,1:]))
data
```

# creating dendrogram

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
```

```python
z=linkage(data,method='ward',metric='euclidean')
plt.figure(figsize=(15,5))
plt.title('hierarchial clustering dendrogram')
plt.xlabel('index')
plt.ylabel('distance')
sch.dendrogram(z)
plt.show()
```

#hierarchical clustering

```python
from sklearn.cluster import AgglomerativeClustering
agglomerative=AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='ward').fit(data)
cluster_labels=pd.Series(agglomerative.labels_)
cluster_labels
```

#concat the clusters to the data

```python
crime['clust']=cluster_labels
crime
```

#groupby  as per labels


```python
crime.iloc[:,1:5].groupby(crime.clust).mean()
```

```python
crime_1=crime[(crime.clust==0)]
crime_1
```

```python
crime_2=crime[(crime.clust==1)]
crime_2
```

```python
crime_3=crime[(crime.clust==2)]
crime_3
```

```python
crime_4=crime[(crime.clust==3)]
crime_4
```

#visualization

```python
crime.plot(x='Rape',y='UrbanPop',c=cluster_labels,kind='scatter',cmap=plt.cm.autumn_r)
plt.title('clustering_formation')
plt.xlabel('Rape')
plt.ylabel('urbanpop')
plt.show()
```

#kmeans clustering

```python
crime_df=crime.iloc[:,1:5] 
crime_df
```

```python
from sklearn.cluster import KMeans
```

```python
fig=plt.figure(figsize=(15,10))
WCSS=[]
for i in range(1,10):
  clf=KMeans(n_clusters=i)
  clf.fit(data)
  WCSS.append(clf.inertia_)
plt.plot(range(1,10),WCSS)
plt.show()
```

```python
crime_kmeans = KMeans(n_clusters = 4).fit(data)
cluster_labels_kmeans=pd.Series(crime_kmeans.labels_)
cluster_labels_kmeans
```

```python
crime_df['clusters']=cluster_labels_kmeans
crime_df
```

#groupby clusters

```python
crime_df.iloc[:,:4].groupby(crime_df.clusters).mean()
```

```python
crime_1=crime_df[(crime_df.clusters==0)]
crime_1
```

```python
crime_2=crime_df[(crime_df.clusters==1)]
crime_2
```

```python
crime_3=crime_df[(crime_df.clusters==2)]
crime_3
```

```python
crime_3=crime_df[(crime_df.clusters==3)]
crime_3
```

```python
crime_df.plot(x='Murder',y='UrbanPop',c=cluster_labels_kmeans,kind='scatter',cmap=plt.cm.gist_heat_r)
```

#dbscan clustering

```python
from sklearn.cluster import DBSCAN
```

```python
crime_data1=pd.read_csv("Crime.csv")
crime_data1
```

```python
crime_dbscan = DBSCAN(eps=0.3,min_samples=5)
crime_dbscan.fit(data) 
```

```python
crime_dbscan.labels_ 
```

```python
crime_labels_dbscan=pd.Series(crime_dbscan.labels_)
crime_labels_dbscan
```

```python
crime_data1['dbscan_cluster']=crime_labels_dbscan
crime_data1 
```

#grouping data as per clusters

```python
crime_data1.groupby(crime_data1.dbscan_cluster).mean()
```

```python
crime_data_1=crime_data1[(crime_data1.dbscan_cluster==-1)]
crime_data_1
```

```python
crime_data_2=crime_data1[(crime_data1.dbscan_cluster==1)]
crime_data_2
```

```python
crime_data_3=crime_data1[(crime_data1.dbscan_cluster==0)]
crime_data_3
```

```python
import sklearn
sklearn.metrics.silhouette_score(data,crime_dbscan.labels_) 
```

```python
from sklearn.neighbors import NearestNeighbors 
```
